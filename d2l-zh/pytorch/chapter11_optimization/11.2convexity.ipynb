### å‡¸é›†

*å‡¸é›†*ï¼ˆconvex setï¼‰æ˜¯å‡¸æ€§çš„åŸºç¡€ã€‚
ç®€å•åœ°è¯´ï¼Œå¦‚æœå¯¹äºä»»ä½•$a, b \in \mathcal{X}$ï¼Œè¿æ¥$a$å’Œ$b$çš„çº¿æ®µä¹Ÿä½äº$\mathcal{X}$ä¸­ï¼Œåˆ™å‘é‡ç©ºé—´ä¸­çš„ä¸€ä¸ªé›†åˆ$\mathcal{X}$æ˜¯*å‡¸*ï¼ˆconvexï¼‰çš„ã€‚
åœ¨æ•°å­¦æœ¯è¯­ä¸Šï¼Œè¿™æ„å‘³ç€å¯¹äºæ‰€æœ‰$\lambda \in [0, 1]$ï¼Œæˆ‘ä»¬å¾—åˆ°

$$\lambda  a + (1-\lambda)  b \in \mathcal{X} \text{ å½“ } a, b \in \mathcal{X}.$$

è¿™å¬èµ·æ¥æœ‰ç‚¹æŠ½è±¡ï¼Œé‚£æˆ‘ä»¬æ¥çœ‹ä¸€ä¸‹ :numref:`fig_pacman`é‡Œçš„ä¾‹å­ã€‚

æ¥ä¸‹æ¥æ¥çœ‹ä¸€ä¸‹äº¤é›† :numref:`fig_convex_intersect`ã€‚
å‡è®¾$\mathcal{X}$å’Œ$\mathcal{Y}$æ˜¯å‡¸é›†ï¼Œé‚£ä¹ˆ$\mathcal {X} \cap \mathcal{Y}$ä¹Ÿæ˜¯å‡¸é›†çš„ã€‚

å› ä¸º$\mathcal{X}$å’Œ$\mathcal{Y}$æ˜¯å‡¸é›†ï¼Œ
æ‰€ä»¥è¿æ¥$a$å’Œ$b$çš„çº¿æ®µåŒ…å«åœ¨$\mathcal{X}$å’Œ$\mathcal{Y}$ä¸­ã€‚
é‰´äºæ­¤ï¼Œå®ƒä»¬ä¹Ÿéœ€è¦åŒ…å«åœ¨$\mathcal {X} \cap \mathcal{Y}$ä¸­ï¼Œä»è€Œè¯æ˜æˆ‘ä»¬çš„å®šç†ã€‚

![ä¸¤ä¸ªå‡¸é›†çš„äº¤é›†æ˜¯å‡¸çš„ã€‚](../img/convex-intersect.svg)

æˆ‘ä»¬å¯ä»¥æ¯«ä¸è´¹åŠ›åœ°è¿›ä¸€æ­¥å¾—åˆ°è¿™æ ·çš„ç»“æœï¼š
ç»™å®šå‡¸é›†$\mathcal{X}_i$ï¼Œå®ƒä»¬çš„äº¤é›†$\cap_{i} \mathcal{X}_i$æ˜¯å‡¸çš„ã€‚
ä½†æ˜¯åå‘æ˜¯ä¸æ­£ç¡®çš„ï¼Œè€ƒè™‘ä¸¤ä¸ªä¸ç›¸äº¤çš„é›†åˆ$\mathcal{X} \cap \mathcal{Y} = \emptyset$ï¼Œ
å–$a \in \mathcal{X}$å’Œ$b \in \mathcal{Y}$ã€‚

é€šå¸¸ï¼Œæ·±åº¦å­¦ä¹ ä¸­çš„é—®é¢˜æ˜¯åœ¨å‡¸é›†ä¸Šå®šä¹‰çš„ã€‚
ä¾‹å¦‚ï¼Œ$\mathbb{R}^d$ï¼Œå³å®æ•°çš„$d$-ç»´å‘é‡çš„é›†åˆæ˜¯å‡¸é›†ï¼ˆæ¯•ç«Ÿ$\mathbb{R}^d$ä¸­ä»»æ„ä¸¤ç‚¹ä¹‹é—´çš„çº¿å­˜åœ¨$\mathbb{R}^d$ï¼‰ä¸­ã€‚
åœ¨æŸäº›æƒ…å†µä¸‹ï¼Œæˆ‘ä»¬ä½¿ç”¨æœ‰ç•Œé•¿åº¦çš„å˜é‡ï¼Œä¾‹å¦‚çƒçš„åŠå¾„å®šä¹‰ä¸º$\{\mathbf{x} | \mathbf{x} \in \mathbb{R}^d \text{ ä¸” } \| \mathbf{x} \| \leq r\}$ã€‚

### å‡¸å‡½æ•°

ç°åœ¨æˆ‘ä»¬æœ‰äº†å‡¸é›†ï¼Œæˆ‘ä»¬å¯ä»¥å¼•å…¥*å‡¸å‡½æ•°*ï¼ˆconvex functionï¼‰$f$ã€‚
ç»™å®šä¸€ä¸ªå‡¸é›†$\mathcal{X}$ï¼Œå¦‚æœå¯¹äºæ‰€æœ‰$x, x' \in \mathcal{X}$å’Œæ‰€æœ‰$\lambda \in [0, 1]$ï¼Œå‡½æ•°$f: \mathcal{X} \to \mathbb{R}$æ˜¯å‡¸çš„ï¼Œæˆ‘ä»¬å¯ä»¥å¾—åˆ°

$$\lambda f(x) + (1-\lambda) f(x') \geq f(\lambda x + (1-\lambda) x').$$

ä¸ºäº†è¯´æ˜è¿™ä¸€ç‚¹ï¼Œè®©æˆ‘ä»¬ç»˜åˆ¶ä¸€äº›å‡½æ•°å¹¶æ£€æŸ¥å“ªäº›å‡½æ•°æ»¡è¶³è¦æ±‚ã€‚
ä¸‹é¢æˆ‘ä»¬å®šä¹‰ä¸€äº›å‡½æ•°ï¼ŒåŒ…æ‹¬å‡¸å‡½æ•°å’Œéå‡¸å‡½æ•°ã€‚


```python
f = lambda x: 0.5 * x**2  # å‡¸å‡½æ•°
g = lambda x: torch.cos(np.pi * x)  # éå‡¸å‡½æ•°
h = lambda x: torch.exp(0.5 * x)  # å‡¸å‡½æ•°

x, segment = torch.arange(-2, 2, 0.01), torch.tensor([-1.5, 1])
d2l.use_svg_display()
_, axes = d2l.plt.subplots(1, 3, figsize=(9, 3))
for ax, func in zip(axes, [f, g, h]):
    d2l.plot([x, segment], [func(x), func(segment)], axes=ax)
```

ä¸å‡ºæ‰€æ–™ï¼Œä½™å¼¦å‡½æ•°ä¸ºéå‡¸çš„ï¼Œè€ŒæŠ›ç‰©çº¿å‡½æ•°å’ŒæŒ‡æ•°å‡½æ•°ä¸ºå‡¸çš„ã€‚
è¯·æ³¨æ„ï¼Œä¸ºä½¿è¯¥æ¡ä»¶æœ‰æ„ä¹‰ï¼Œ$\mathcal{X}$æ˜¯å‡¸é›†çš„è¦æ±‚æ˜¯å¿…è¦çš„ã€‚
å¦åˆ™å¯èƒ½æ— æ³•å¾ˆå¥½åœ°ç•Œå®š$f(\lambda x + (1-\lambda) x')$çš„ç»“æœã€‚

### è©¹æ£®ä¸ç­‰å¼

ç»™å®šä¸€ä¸ªå‡¸å‡½æ•°$f$ï¼Œæœ€æœ‰ç”¨çš„æ•°å­¦å·¥å…·ä¹‹ä¸€å°±æ˜¯*è©¹æ£®ä¸ç­‰å¼*ï¼ˆJensen's inequalityï¼‰ã€‚
å®ƒæ˜¯å‡¸æ€§å®šä¹‰çš„ä¸€ç§æ¨å¹¿ï¼š

$$\sum_i \alpha_i f(x_i)  \geq f\left(\sum_i \alpha_i x_i\right) \text{ and } E_X[f(X)] \geq f\left(E_X[X]\right),$$
:eqlabel:`eq_jensens-inequality`

å…¶ä¸­$\alpha_i$æ˜¯æ»¡è¶³$\sum_i \alpha_i = 1$çš„éè´Ÿå®æ•°ï¼Œ$X$æ˜¯éšæœºå˜é‡ã€‚
æ¢å¥è¯è¯´ï¼Œå‡¸å‡½æ•°çš„æœŸæœ›ä¸å°äºæœŸæœ›çš„å‡¸å‡½æ•°ï¼Œå…¶ä¸­åè€…é€šå¸¸æ˜¯ä¸€ä¸ªæ›´ç®€å•çš„è¡¨è¾¾å¼ã€‚
ä¸ºäº†è¯æ˜ç¬¬ä¸€ä¸ªä¸ç­‰å¼ï¼Œæˆ‘ä»¬å¤šæ¬¡å°†å‡¸æ€§çš„å®šä¹‰åº”ç”¨äºä¸€æ¬¡æ±‚å’Œä¸­çš„ä¸€é¡¹ã€‚

è©¹æ£®ä¸ç­‰å¼çš„ä¸€ä¸ªå¸¸è§åº”ç”¨ï¼šç”¨ä¸€ä¸ªè¾ƒç®€å•çš„è¡¨è¾¾å¼çº¦æŸä¸€ä¸ªè¾ƒå¤æ‚çš„è¡¨è¾¾å¼ã€‚
ä¾‹å¦‚ï¼Œå®ƒå¯ä»¥åº”ç”¨äºéƒ¨åˆ†è§‚å¯Ÿåˆ°çš„éšæœºå˜é‡çš„å¯¹æ•°ä¼¼ç„¶ã€‚
å…·ä½“åœ°è¯´ï¼Œç”±äº$\int P(Y) P(X \mid Y) dY = P(X)$ï¼Œæ‰€ä»¥

$$E_{Y \sim P(Y)}[-\log P(X \mid Y)] \geq -\log P(X),$$

è¿™é‡Œï¼Œ$Y$æ˜¯å…¸å‹çš„æœªè§‚å¯Ÿåˆ°çš„éšæœºå˜é‡ï¼Œ$P(Y)$æ˜¯å®ƒå¯èƒ½å¦‚ä½•åˆ†å¸ƒçš„æœ€ä½³çŒœæµ‹ï¼Œ$P(X)$æ˜¯å°†$Y$ç§¯åˆ†åçš„åˆ†å¸ƒã€‚
ä¾‹å¦‚ï¼Œåœ¨èšç±»ä¸­$Y$å¯èƒ½æ˜¯ç°‡æ ‡ç­¾ï¼Œè€Œåœ¨åº”ç”¨ç°‡æ ‡ç­¾æ—¶ï¼Œ$P(X \mid Y)$æ˜¯ç”Ÿæˆæ¨¡å‹ã€‚

## æ€§è´¨

ä¸‹é¢æˆ‘ä»¬æ¥çœ‹ä¸€ä¸‹å‡¸å‡½æ•°ä¸€äº›æœ‰è¶£çš„æ€§è´¨ã€‚

### å±€éƒ¨æå°å€¼æ˜¯å…¨å±€æå°å€¼

é¦–å…ˆå‡¸å‡½æ•°çš„å±€éƒ¨æå°å€¼ä¹Ÿæ˜¯å…¨å±€æå°å€¼ã€‚
ä¸‹é¢æˆ‘ä»¬ç”¨åè¯æ³•ç»™å‡ºè¯æ˜ã€‚

å‡è®¾$x^{\ast} \in \mathcal{X}$æ˜¯ä¸€ä¸ªå±€éƒ¨æœ€å°å€¼ï¼Œåˆ™å­˜åœ¨ä¸€ä¸ªå¾ˆå°çš„æ­£å€¼$p$ï¼Œä½¿å¾—å½“$x \in \mathcal{X}$æ»¡è¶³$0 < |x - x^{\ast}| \leq p$æ—¶ï¼Œæœ‰$f(x^{\ast}) < f(x)$ã€‚

ç°åœ¨å‡è®¾å±€éƒ¨æå°å€¼$x^{\ast}$ä¸æ˜¯$f$çš„å…¨å±€æå°å€¼ï¼šå­˜åœ¨$x' \in \mathcal{X}$ä½¿å¾—$f(x') < f(x^{\ast})$ã€‚
åˆ™å­˜åœ¨
$\lambda \in [0, 1)$ï¼Œæ¯”å¦‚$\lambda = 1 - \frac{p}{|x^{\ast} - x'|}$ï¼Œä½¿å¾—
$0 < |\lambda x^{\ast} + (1-\lambda) x' - x^{\ast}| \leq p$ã€‚

ç„¶è€Œï¼Œæ ¹æ®å‡¸æ€§çš„æ€§è´¨ï¼Œæœ‰

$$\begin{aligned}
    f(\lambda x^{\ast} + (1-\lambda) x') &\leq \lambda f(x^{\ast}) + (1-\lambda) f(x') \\
    &< \lambda f(x^{\ast}) + (1-\lambda) f(x^{\ast}) \\

\end{aligned}$$

è¿™ä¸$x^{\ast}$æ˜¯å±€éƒ¨æœ€å°å€¼ç›¸çŸ›ç›¾ã€‚
å› æ­¤ï¼Œä¸å­˜åœ¨$x' \in \mathcal{X}$æ»¡è¶³$f(x') < f(x^{\ast})$ã€‚
ç»¼ä¸Šæ‰€è¿°ï¼Œå±€éƒ¨æœ€å°å€¼$x^{\ast}$ä¹Ÿæ˜¯å…¨å±€æœ€å°å€¼ã€‚

ä¾‹å¦‚ï¼Œå¯¹äºå‡¸å‡½æ•°$f(x) = (x-1)^2$ï¼Œæœ‰ä¸€ä¸ªå±€éƒ¨æœ€å°å€¼$x=1$ï¼Œå®ƒä¹Ÿæ˜¯å…¨å±€æœ€å°å€¼ã€‚


ä½ è¿™å¼ å›¾å±•ç¤ºçš„æ˜¯ **ç¬¬11.1.3èŠ‚ï¼šå‡¸æ€§ï¼ˆConvexityï¼‰** çš„è¿›ä¸€æ­¥æ•°å­¦åŸºç¡€ï¼Œä¸»è¦åŒ…æ‹¬ä¸¤ä¸ªå…³é”®çŸ¥è¯†ç‚¹ï¼š

## ğŸ§® ä¸€ã€**è©¹æ£®ä¸ç­‰å¼ï¼ˆJensen's Inequalityï¼‰**

### ğŸ“Œ æ˜¯ä»€ä¹ˆï¼Ÿ

è©¹æ£®ä¸ç­‰å¼æ˜¯å‡¸å‡½æ•°æ€§è´¨çš„ä¸€ä¸ªé‡è¦æ¨å¹¿ï¼š

$$
f\left(\sum_i \alpha_i x_i \right) \leq \sum_i \alpha_i f(x_i) \quad \text{ï¼ˆå‡¸å‡½æ•°æˆç«‹ï¼‰}
$$

æˆ–å†™æˆæœŸæœ›å½¢å¼ï¼š

$$

### ğŸ§  å¦‚ä½•ç†è§£ï¼Ÿ

> å¦‚æœå‡½æ•°æ˜¯å‡¸çš„ï¼Œé‚£ä¹ˆâ€œå…ˆå–å¹³å‡å†è®¡ç®—å‡½æ•°å€¼â€å¾—åˆ°çš„ç»“æœ**ä¸å¦‚**â€œå…ˆåˆ†åˆ«ç®—å‡½æ•°å€¼å†å¹³å‡â€ã€‚

### ğŸ“Œ æœ‰ä»€ä¹ˆç”¨ï¼Ÿ

1. æ˜¯è¯æ˜å¾ˆå¤šæ·±åº¦å­¦ä¹ ä¼˜åŒ–ä¸ç­‰å¼çš„åŸºç¡€ï¼ˆæ¯”å¦‚ KL æ•£åº¦æ€»æ˜¯éè´Ÿï¼‰ã€‚
2. åœ¨ç”Ÿæˆæ¨¡å‹ä¸­ï¼Œå¸¸ç”¨äºå¤„ç†éšå˜é‡çš„è¾¹é™…å¯¹æ•°ä¼¼ç„¶ï¼ˆå¦‚å˜åˆ†æ¨æ–­ä¸­ä½¿ç”¨ï¼‰ã€‚

### ğŸ“Œ å›¾ä¸­åº”ç”¨ä¸¾ä¾‹ï¼š

å‡è®¾æˆ‘ä»¬è¦æœ€å¤§åŒ–è¾¹é™…æ¦‚ç‡ï¼š

$$

æˆ‘ä»¬æ— æ³•ç›´æ¥è®¡ç®—ï¼Œä½†å¯ä»¥é€šè¿‡ Jensen ä¸ç­‰å¼ï¼š

$$

## ğŸ§  äºŒã€å‡¸å‡½æ•°çš„ä¸€ä¸ªé‡è¦æ€§è´¨ï¼š**å±€éƒ¨æœ€å° = å…¨å±€æœ€å°**

4. è‹¥ $f(x') < f(x^*)$ï¼Œåˆ™ä¼šå¯¼è‡´æ„é€ å‡ºçš„ç‚¹å‡½æ•°å€¼ä¹Ÿå°äº $f(x^*)$ï¼Œè¿™ä¸â€œå±€éƒ¨æœ€å°â€çŸ›ç›¾ã€‚

å‡¸å‡½æ•°çš„å±€éƒ¨æå°å€¼åŒæ—¶ä¹Ÿæ˜¯å…¨å±€æå°å€¼è¿™ä¸€æ€§è´¨æ˜¯å¾ˆæ–¹ä¾¿çš„ã€‚
è¿™æ„å‘³ç€å¦‚æœæˆ‘ä»¬æœ€å°åŒ–å‡½æ•°ï¼Œæˆ‘ä»¬å°±ä¸ä¼šâ€œå¡ä½â€ã€‚
