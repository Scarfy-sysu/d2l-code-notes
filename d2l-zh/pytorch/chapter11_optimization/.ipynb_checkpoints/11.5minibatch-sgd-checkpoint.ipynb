## ğŸ“Œ å›¾ç¤ºï¼šå°æ‰¹é‡å’Œå‘é‡åŒ–çš„åŠ é€ŸæŠ€å·§

è¯¾æœ¬ä¸­ä¸¾äº†çŸ©é˜µä¹˜æ³• `A * B = C` çš„ä¾‹å­ã€‚ä¼˜åŒ–æ–¹æ³•åŒ…æ‹¬ï¼š

1. **æŒ‰åˆ—ä¹˜æ³•**ï¼ˆæ¯æ¬¡è®¡ç®— C çš„ä¸€åˆ—ï¼‰ï¼›
2. **æŒ‰è¡Œä¹˜æ³•**ï¼ˆæ¯æ¬¡è®¡ç®— A çš„ä¸€è¡Œï¼‰ï¼›
3. **å—çŸ©é˜µä¹˜æ³•ï¼ˆtile-basedï¼‰**ï¼ˆä¸€æ¬¡è®¡ç®—ä¸€ä¸ªå°åŒºå—ï¼‰ï¼›
4. **ä¿æŒè®¿é—®é¡ºåºä¸€è‡´æ€§**ï¼Œå‡å°‘å†…å­˜è·³è½¬ï¼›

è¿™äº›ç­–ç•¥é…åˆå°æ‰¹é‡ SGD å¯æå¤§æå‡è®­ç»ƒæ•ˆç‡ã€‚

## ğŸ§ª å°ç»“

| æ–¹æ³•        | ç‰¹ç‚¹         | é—®é¢˜              |
| --------- | ---------- | --------------- |
| å®Œæ•´æ¢¯åº¦ä¸‹é™    | æ›´æ–°å‡†ç¡®       | å¤ªæ…¢ï¼ŒGPU ä¸å‹å¥½      |
| éšæœºæ¢¯åº¦ä¸‹é™    | å¿«ï¼Œèµ„æºå°‘      | å™ªå£°å¤§ï¼Œæ”¶æ•›ä¸ç¨³å®š       |
| å°æ‰¹é‡æ¢¯åº¦ä¸‹é™ âœ… | å¿«é€Ÿã€ç¨³å®šã€ç¡¬ä»¶å‹å¥½ | éœ€è¦è°ƒå¥½ batch size |

````python
%matplotlib inline
import numpy as np
import torch
from torch import nn
from d2l import torch as d2l

timer = d2l.Timer()
A = torch.zeros(256, 256)
B = torch.randn(256, 256)
C = torch.randn(256, 256)
````

æŒ‰å…ƒç´ åˆ†é…åªéœ€éå†åˆ†åˆ«ä¸º$\mathbf{B}$å’Œ$\mathbf{C}$çš„æ‰€æœ‰è¡Œå’Œåˆ—ï¼Œå³å¯å°†è¯¥å€¼åˆ†é…ç»™$\mathbf{A}$ã€‚


````python
# é€å…ƒç´ è®¡ç®—A=BC
timer.start()
for i in range(256):
    for j in range(256):
        A[i, j] = torch.dot(B[i, :], C[:, j])
timer.stop()
````

æ›´å¿«çš„ç­–ç•¥æ˜¯æ‰§è¡ŒæŒ‰åˆ—åˆ†é…ã€‚


````python
# é€åˆ—è®¡ç®—A=BC
timer.start()
for j in range(256):
    A[:, j] = torch.mv(B, C[:, j])
timer.stop()
````

æœ€æœ‰æ•ˆçš„æ–¹æ³•æ˜¯åœ¨ä¸€ä¸ªåŒºå—ä¸­æ‰§è¡Œæ•´ä¸ªæ“ä½œã€‚è®©æˆ‘ä»¬çœ‹çœ‹å®ƒä»¬å„è‡ªçš„æ“ä½œé€Ÿåº¦æ˜¯å¤šå°‘ã€‚


````python
# ä¸€æ¬¡æ€§è®¡ç®—A=BC
timer.start()
A = torch.mm(B, C)
timer.stop()

# ä¹˜æ³•å’ŒåŠ æ³•ä½œä¸ºå•ç‹¬çš„æ“ä½œï¼ˆåœ¨å®è·µä¸­èåˆï¼‰
gigaflops = [2/i for i in timer.times]
print(f'performance in Gigaflops: element {gigaflops[0]:.3f}, '
      f'column {gigaflops[1]:.3f}, full {gigaflops[2]:.3f}')
````

## å°æ‰¹é‡

:label:`sec_minibatches`

ä¹‹å‰æˆ‘ä»¬ä¼šç†æ‰€å½“ç„¶åœ°è¯»å–æ•°æ®çš„*å°æ‰¹é‡*ï¼Œè€Œä¸æ˜¯è§‚æµ‹å•ä¸ªæ•°æ®æ¥æ›´æ–°å‚æ•°ï¼Œç°åœ¨ç®€è¦è§£é‡Šä¸€ä¸‹åŸå› ã€‚
å¤„ç†å•ä¸ªè§‚æµ‹å€¼éœ€è¦æˆ‘ä»¬æ‰§è¡Œè®¸å¤šå•ä¸€çŸ©é˜µ-çŸ¢é‡ï¼ˆç”šè‡³çŸ¢é‡-çŸ¢é‡ï¼‰ä¹˜æ³•ï¼Œè¿™è€—è´¹ç›¸å½“å¤§ï¼Œè€Œä¸”å¯¹åº”æ·±åº¦å­¦ä¹ æ¡†æ¶ä¹Ÿè¦å·¨å¤§çš„å¼€é”€ã€‚
è¿™æ—¢é€‚ç”¨äºè®¡ç®—æ¢¯åº¦ä»¥æ›´æ–°å‚æ•°æ—¶ï¼Œä¹Ÿé€‚ç”¨äºç”¨ç¥ç»ç½‘ç»œé¢„æµ‹ã€‚
ä¹Ÿå°±æ˜¯è¯´ï¼Œæ¯å½“æˆ‘ä»¬æ‰§è¡Œ$\mathbf{w} \leftarrow \mathbf{w} - \eta_t \mathbf{g}_t$æ—¶ï¼Œæ¶ˆè€—å·¨å¤§ã€‚å…¶ä¸­

$$\mathbf{g}_t = \partial_{\mathbf{w}} f(\mathbf{x}_{t}, \mathbf{w}).$$

æˆ‘ä»¬å¯ä»¥é€šè¿‡å°†å…¶åº”ç”¨äºä¸€ä¸ªå°æ‰¹é‡è§‚æµ‹å€¼æ¥æé«˜æ­¤æ“ä½œçš„*è®¡ç®—*æ•ˆç‡ã€‚
ä¹Ÿå°±æ˜¯è¯´ï¼Œæˆ‘ä»¬å°†æ¢¯åº¦$\mathbf{g}_t$æ›¿æ¢ä¸ºä¸€ä¸ªå°æ‰¹é‡è€Œä¸æ˜¯å•ä¸ªè§‚æµ‹å€¼

$$\mathbf{g}_t = \partial_{\mathbf{w}} \frac{1}{|\mathcal{B}_t|} \sum_{i \in \mathcal{B}_t} f(\mathbf{x}_{i}, \mathbf{w}).$$

è®©æˆ‘ä»¬çœ‹çœ‹è¿™å¯¹$\mathbf{g}_t$çš„ç»Ÿè®¡å±æ€§æœ‰ä»€ä¹ˆå½±å“ï¼šç”±äº$\mathbf{x}_t$å’Œå°æ‰¹é‡$\mathcal{B}_t$çš„æ‰€æœ‰å…ƒç´ éƒ½æ˜¯ä»è®­ç»ƒé›†ä¸­éšæœºæŠ½å‡ºçš„ï¼Œå› æ­¤æ¢¯åº¦çš„æœŸæœ›ä¿æŒä¸å˜ã€‚
å¦ä¸€æ–¹é¢ï¼Œæ–¹å·®æ˜¾è‘—é™ä½ã€‚
ç”±äºå°æ‰¹é‡æ¢¯åº¦ç”±æ­£åœ¨è¢«å¹³å‡è®¡ç®—çš„$b := |\mathcal{B}_t|$ä¸ªç‹¬ç«‹æ¢¯åº¦ç»„æˆï¼Œå…¶æ ‡å‡†å·®é™ä½äº†$b^{-\frac{1}{2}}$ã€‚
è¿™æœ¬èº«å°±æ˜¯ä¸€ä»¶å¥½äº‹ï¼Œå› ä¸ºè¿™æ„å‘³ç€æ›´æ–°ä¸å®Œæ•´çš„æ¢¯åº¦æ›´æ¥è¿‘äº†ã€‚

ç›´è§‚æ¥è¯´ï¼Œè¿™è¡¨æ˜é€‰æ‹©å¤§å‹çš„å°æ‰¹é‡$\mathcal{B}_t$å°†æ˜¯æ™®éå¯è¡Œçš„ã€‚
ç„¶è€Œï¼Œç»è¿‡ä¸€æ®µæ—¶é—´åï¼Œä¸è®¡ç®—ä»£ä»·çš„çº¿æ€§å¢é•¿ç›¸æ¯”ï¼Œæ ‡å‡†å·®çš„é¢å¤–å‡å°‘æ˜¯å¾®ä¹å…¶å¾®çš„ã€‚
åœ¨å®è·µä¸­æˆ‘ä»¬é€‰æ‹©ä¸€ä¸ªè¶³å¤Ÿå¤§çš„å°æ‰¹é‡ï¼Œå®ƒå¯ä»¥æä¾›è‰¯å¥½çš„è®¡ç®—æ•ˆç‡åŒæ—¶ä»é€‚åˆGPUçš„å†…å­˜ã€‚
ä¸‹é¢ï¼Œæˆ‘ä»¬æ¥çœ‹çœ‹è¿™äº›é«˜æ•ˆçš„ä»£ç ã€‚
åœ¨é‡Œé¢æˆ‘ä»¬æ‰§è¡Œç›¸åŒçš„çŸ©é˜µ-çŸ©é˜µä¹˜æ³•ï¼Œä½†æ˜¯è¿™æ¬¡æˆ‘ä»¬å°†å…¶ä¸€æ¬¡æ€§åˆ†ä¸º64åˆ—çš„â€œå°æ‰¹é‡â€ã€‚


````python
timer.start()
for j in range(0, 256, 64):
    A[:, j:j+64] = torch.mm(B, C[:, j:j+64])
timer.stop()
print(f'performance in Gigaflops: block {2 / timer.times[3]:.3f}')
````

## è¯»å–æ•°æ®é›†

è®©æˆ‘ä»¬æ¥çœ‹çœ‹å¦‚ä½•ä»æ•°æ®ä¸­æœ‰æ•ˆåœ°ç”Ÿæˆå°æ‰¹é‡ã€‚
ä¸‹é¢æˆ‘ä»¬ä½¿ç”¨NASAå¼€å‘çš„æµ‹è¯•æœºç¿¼çš„æ•°æ®é›†[ä¸åŒé£è¡Œå™¨äº§ç”Ÿçš„å™ªå£°](https://archive.ics.uci.edu/ml/datasets/Airfoil+Self-Noise)æ¥æ¯”è¾ƒè¿™äº›ä¼˜åŒ–ç®—æ³•ã€‚
ä¸ºæ–¹ä¾¿èµ·è§ï¼Œæˆ‘ä»¬åªä½¿ç”¨å‰$1,500$æ ·æœ¬ã€‚
æ•°æ®å·²ä½œé¢„å¤„ç†ï¼šæˆ‘ä»¬ç§»é™¤äº†å‡å€¼å¹¶å°†æ–¹å·®é‡æ–°ç¼©æ”¾åˆ°æ¯ä¸ªåæ ‡ä¸º$1$ã€‚

````python
# ä» NASA çš„æ°”æµå™ªå£°å®éªŒæ•°æ®ä¸­åŠ è½½æ•°æ®ï¼Œå¹¶åˆ’åˆ†æˆå°æ‰¹é‡ä¾›è®­ç»ƒä½¿ç”¨ã€‚

#@save
d2l.DATA_HUB['airfoil'] = (d2l.DATA_URL + 'airfoil_self_noise.dat',
                           '76e5be1548fd8222e5074cf0faae75edff8cf93f')

#@save
def get_data_ch11(batch_size=10, n=1500):
    # ä»txtè¯»å–1500ä¸ªæ ·æœ¬
    data = np.genfromtxt(d2l.download('airfoil'),
                         dtype=np.float32, delimiter='\t')
    
    # æ ‡å‡†åŒ–ï¼šå‡å»å‡å€¼é™¤ä»¥æ ‡å‡†å·®
    data = torch.from_numpy((data - data.mean(axis=0)) / data.std(axis=0))
    data_iter = d2l.load_array((data[:n, :-1], data[:n, -1]),
                               batch_size, is_train=True)
    
    # data[:, :-1] æ˜¯æ‰€æœ‰ç‰¹å¾ï¼ˆæ¯è¡Œæœ€åä¸€åˆ—å‰çš„æ‰€æœ‰æ•°æ®ï¼‰ï¼›
    # data[:, -1] æ˜¯æ ‡ç­¾ï¼ˆæœ€åä¸€åˆ—ï¼‰ï¼›
    # load_array(...) ä¼šå°†æ•´ä¸ªæ•°æ®é›†åˆ’åˆ†æˆä¸€ä¸ªä¸ª batchï¼Œè¿”å›å¯ä»¥è¿­ä»£çš„å°æ‰¹é‡è®­ç»ƒæ•°æ®ï¼ˆdata_iterï¼‰ã€‚
    return data_iter, data.shape[1]-1
````

## ä»é›¶å¼€å§‹å®ç°

 :numref:`sec_linear_scratch`ä¸€èŠ‚ä¸­å·²ç»å®ç°è¿‡å°æ‰¹é‡éšæœºæ¢¯åº¦ä¸‹é™ç®—æ³•ã€‚
æˆ‘ä»¬åœ¨è¿™é‡Œå°†å®ƒçš„è¾“å…¥å‚æ•°å˜å¾—æ›´åŠ é€šç”¨ï¼Œä¸»è¦æ˜¯ä¸ºäº†æ–¹ä¾¿æœ¬ç« åé¢ä»‹ç»çš„å…¶ä»–ä¼˜åŒ–ç®—æ³•ä¹Ÿå¯ä»¥ä½¿ç”¨åŒæ ·çš„è¾“å…¥ã€‚
å…·ä½“æ¥è¯´ï¼Œæˆ‘ä»¬æ·»åŠ äº†ä¸€ä¸ªçŠ¶æ€è¾“å…¥`states`å¹¶å°†è¶…å‚æ•°æ”¾åœ¨å­—å…¸`hyperparams`ä¸­ã€‚
æ­¤å¤–ï¼Œæˆ‘ä»¬å°†åœ¨è®­ç»ƒå‡½æ•°é‡Œå¯¹å„ä¸ªå°æ‰¹é‡æ ·æœ¬çš„æŸå¤±æ±‚å¹³å‡ï¼Œå› æ­¤ä¼˜åŒ–ç®—æ³•ä¸­çš„æ¢¯åº¦ä¸éœ€è¦é™¤ä»¥æ‰¹é‡å¤§å°ã€‚


````python
def sgd(params, states, hyperparams):
    for p in params:
        p.data.sub_(hyperparams['lr'] * p.grad)
        p.grad.data.zero_()
````

ä¸‹é¢å®ç°ä¸€ä¸ªé€šç”¨çš„è®­ç»ƒå‡½æ•°ï¼Œä»¥æ–¹ä¾¿æœ¬ç« åé¢ä»‹ç»çš„å…¶ä»–ä¼˜åŒ–ç®—æ³•ä½¿ç”¨ã€‚
å®ƒåˆå§‹åŒ–äº†ä¸€ä¸ªçº¿æ€§å›å½’æ¨¡å‹ï¼Œç„¶åå¯ä»¥ä½¿ç”¨å°æ‰¹é‡éšæœºæ¢¯åº¦ä¸‹é™ä»¥åŠåç»­å°èŠ‚ä»‹ç»çš„å…¶ä»–ç®—æ³•æ¥è®­ç»ƒæ¨¡å‹ã€‚


````python
#@save
def train_ch11(trainer_fn, states, hyperparams, data_iter,
               feature_dim, num_epochs=2):
    # åˆå§‹åŒ–æ¨¡å‹
    w = torch.normal(mean=0.0, std=0.01, size=(feature_dim, 1),
                     requires_grad=True)
    b = torch.zeros((1), requires_grad=True)
    net, loss = lambda X: d2l.linreg(X, w, b), d2l.squared_loss
    # è®­ç»ƒæ¨¡å‹
    animator = d2l.Animator(xlabel='epoch', ylabel='loss',
                            xlim=[0, num_epochs], ylim=[0.22, 0.35])
    n, timer = 0, d2l.Timer()
    for _ in range(num_epochs):
        for X, y in data_iter:
            l = loss(net(X), y).mean()
            l.backward()
            trainer_fn([w, b], states, hyperparams)
            n += X.shape[0]
            if n % 200 == 0:
                timer.stop()
                animator.add(n/X.shape[0]/len(data_iter),
                             (d2l.evaluate_loss(net, data_iter, loss),))
                timer.start()
    print(f'loss: {animator.Y[0][-1]:.3f}, {timer.avg():.3f} sec/epoch')
    return timer.cumsum(), animator.Y[0]
````

è®©æˆ‘ä»¬æ¥çœ‹çœ‹æ‰¹é‡æ¢¯åº¦ä¸‹é™çš„ä¼˜åŒ–æ˜¯å¦‚ä½•è¿›è¡Œçš„ã€‚
è¿™å¯ä»¥é€šè¿‡å°†å°æ‰¹é‡è®¾ç½®ä¸º1500ï¼ˆå³æ ·æœ¬æ€»æ•°ï¼‰æ¥å®ç°ã€‚
å› æ­¤ï¼Œæ¨¡å‹å‚æ•°æ¯ä¸ªè¿­ä»£è½®æ•°åªè¿­ä»£ä¸€æ¬¡ã€‚


````python
def train_sgd(lr, batch_size, num_epochs=2):
    data_iter, feature_dim = get_data_ch11(batch_size)
    return train_ch11(
        sgd, None, {'lr': lr}, data_iter, feature_dim, num_epochs)

gd_res = train_sgd(1, 1500, 10)
````

ä¸ºäº†ç®€åŒ–å®ç°ï¼Œæˆ‘ä»¬é€‰æ‹©äº†å¾ˆå°çš„å­¦ä¹ ç‡ã€‚
åœ¨éšæœºæ¢¯åº¦ä¸‹é™çš„å®éªŒä¸­ï¼Œæ¯å½“ä¸€ä¸ªæ ·æœ¬è¢«å¤„ç†ï¼Œæ¨¡å‹å‚æ•°éƒ½ä¼šæ›´æ–°ã€‚
åœ¨è¿™ä¸ªä¾‹å­ä¸­ï¼Œè¿™ç›¸å½“äºæ¯ä¸ªè¿­ä»£è½®æ•°æœ‰1500æ¬¡æ›´æ–°ã€‚

## ç®€æ´å®ç°

ä¸‹é¢ç”¨æ·±åº¦å­¦ä¹ æ¡†æ¶è‡ªå¸¦ç®—æ³•å®ç°ä¸€ä¸ªé€šç”¨çš„è®­ç»ƒå‡½æ•°ï¼Œæˆ‘ä»¬å°†åœ¨æœ¬ç« ä¸­å…¶å®ƒå°èŠ‚ä½¿ç”¨å®ƒã€‚


````python
#@save
def train_concise_ch11(trainer_fn, hyperparams, data_iter, num_epochs=4):
    # åˆå§‹åŒ–æ¨¡å‹
    net = nn.Sequential(nn.Linear(5, 1))
    def init_weights(m):
        if type(m) == nn.Linear:
            torch.nn.init.normal_(m.weight, std=0.01)
    net.apply(init_weights)

    optimizer = trainer_fn(net.parameters(), **hyperparams)
    loss = nn.MSELoss(reduction='none')
    animator = d2l.Animator(xlabel='epoch', ylabel='loss',
                            xlim=[0, num_epochs], ylim=[0.22, 0.35])
    n, timer = 0, d2l.Timer()
    for _ in range(num_epochs):
        for X, y in data_iter:
            optimizer.zero_grad()
            out = net(X)
            y = y.reshape(out.shape)
            l = loss(out, y)
            l.mean().backward()
            optimizer.step()
            n += X.shape[0]
            if n % 200 == 0:
                timer.stop()
                # MSELossè®¡ç®—å¹³æ–¹è¯¯å·®æ—¶ä¸å¸¦ç³»æ•°1/2
                animator.add(n/X.shape[0]/len(data_iter),
                             (d2l.evaluate_loss(net, data_iter, loss) / 2,))
                timer.start()
    print(f'loss: {animator.Y[0][-1]:.3f}, {timer.avg():.3f} sec/epoch')
````

ä¸‹é¢ä½¿ç”¨è¿™ä¸ªè®­ç»ƒå‡½æ•°ï¼Œå¤ç°ä¹‹å‰çš„å®éªŒã€‚


````python
data_iter, _ = get_data_ch11(10)
trainer = torch.optim.SGD
train_concise_ch11(trainer, {'lr': 0.01}, data_iter)
````
